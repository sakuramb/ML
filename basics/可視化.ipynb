```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(100)

# 一様乱数を10,000個生成
random_number_1 = np.random.rand(10000)

# 正規分布に従う乱数を10,000個生成
random_number_2 = np.random.randn(10000)
# 二項分布に従う乱数を 10,000個生成,成功確率は0.5
random_number_3 = np.random.binomial(100, 0.5, size=(10000))


# 一様乱数をヒストグラムで表示。binsは50に指定
plt.figure(figsize=(2,2))
plt.hist(random_number_1, bins=50)
plt.title('uniform_distribution')
plt.grid(True)
plt.show()

# 正規分布に従う乱数をヒストグラムで表示
plt.hist(random_number_2, bins=50)
plt.title('normal_distribution')
plt.grid(True)
plt.show()

# 二項分布に従う乱数をヒストグラムで表示。binsは50に指定
plt.hist(random_number_3, bins=50)
plt.title('binomial_distribution')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 5*np.pi)
y = np.sin(x)
plt.plot(x, y)
# plt.ylim([0, 1])
# plt.xlim([0, 10])
plt.title("y=sin(x)( 0< y< 1)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)
labels = ["90°", "180°", "270°", "360°"]
positions = [np.pi/2, np.pi, np.pi*3/2, np.pi*2]
plt.xticks(positions, labels)
plt.plot(x, y1, color='b', label='y=sin(x)')
plt.plot(x, y2, color='k', label='y=cos(x)')
plt.legend(["y=sin(x)", "y=cos(x)"])
plt.show()
```
